```python
Options = ["Balance", "Withdraw", "Change PIN", "Exit"]

print("----Welcome To ATM Services----")
print("Please Insert Your Card")
print("Your Card is Inserted")

if input("Please Enter Your Pin: ") == "1234":
    print("Your Pin is Accepted")
    for I in Options:
        print(I)

    SERVICES = int(input("Please Select Your Option: "))
    if SERVICES == 1:
        print("Your Balance is: 12")
    elif SERVICES == 2:
        WITHDRAW = int(input("Please Enter The Amount You Want To Withdraw: "))
        print(f"You Have Withdrawn: {WITHDRAW}")
    elif SERVICES == 3:
        NEW_PIN = input("Please Enter Your New Pin: ")
        print(f"Your Pin Has Been Changed To: {NEW_PIN}")
    elif SERVICES == 4:
        print("Thank You For Using Our Services")
    else:
        print("Invalid Option Selected")
else:
    print("Invalid Pin")
```
